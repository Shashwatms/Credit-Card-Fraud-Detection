Under-Sampling (Method Here we'll use to deal with imbalance data)



Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent Transactions --> 492

We'll take a random of 492 sample data from legit data or normal transaction(284315) an then concatenate/join with fraud data (492). Once we do that we'll 492 + 492 Data (legit + fraud).It will be very good dataset with uniformm distribution. Then we can make better prediction with machine learning.

```python
legit_sample = legit.sample(n=492)
```

Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)  #concatenating rowise
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Here mean value almost same to prv one example V1 has almost same value.It means we got good sample here.The nature of
#dataset has not changed.For bad sample mean value would have been very different.
```

Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Class', axis=1)  #removing class from x variable. (Features)
Y = new_dataset['Class']                       # storing class into y variable.  (Targets)
```

```python
X
```

```python
print(Y)
```

Split the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2) #stratify tells even distribution
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression :

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data

model.fit(X_train, Y_train)
```

Model Evaluation :

Accuracy Score

```python
# accuracy on training data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
